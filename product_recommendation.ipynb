# Reversing the map of indices and product

```python
product_indexes= pd.Series(df.index, index= df['product_name']).drop_duplicates()
product_indexes
```

```python
# Importing Linear Kernel
from sklearn.metrics.pairwise import linear_kernel


# Function that takes in product title as input and outputs most similar product

def recommend_product(text):
    index= product_indexes[text]
    
    cosine_sim= linear_kernel(tfidf_matrix[index], tfidf_matrix)
    sim_scores= list(enumerate(cosine_sim[0]))
    
    sim_scores= sorted(sim_scores, key= lambda x: x[1], reverse= True)
    
    sim_scores= sim_scores[1:10]
    
    product_indices= [i[0] for i in sim_scores]
    
    return df['product_name'].iloc[product_indices]
```

```python
recommended_product= recommend_product(input("Enter a product name: "))
if recommended_product is not None: 
    print("Similar Products")
    print("\n")
    for product_name in recommended_product:
        print(product_name)
```

# **Product Search Engine**

The problem is on optimizing the search engine to get better search result. 
The biggest challenge is that most search engines are string matching. But search should not work on the based on the string match, it should also consider other words.

One of the way to solve this problem is word embeddings. word2vec is one of
methods to construct such an embedding. The two ways the embeddings can be 
built: skip- gram and common bag of words.

The CBOW method takes each word's context as input and predicts the word corresponding to the context. The input to the network context and passed to the N neurons and then at the end, the output layer predicts the word using softmax layer.

```python
import numpy as np
import re
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from scipy import spatial
from gensim.models import KeyedVectors

# Download required NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Load stopwords
stop = set(stopwords.words('english'))

# Define the remove_stopwords function
def remove_stopwords(text, is_lower_case=False):
    if not text:
        return ""
    pattern = r'[^a-zA-Z0-9\s]'
    text = re.sub(pattern, ' ', text)
    tokens = nltk.word_tokenize(text)
    tokens = [token.strip() for token in tokens]
    if is_lower_case:
        filtered_tokens = [token for token in tokens if token not in stop]
    else:
        filtered_tokens = [token for token in tokens if token.lower() not in stop]
    filtered_text = ' '.join(filtered_tokens)
    return filtered_text

# Define the get_embedding function
def get_embedding(word):
    if word in model.key_to_index:
        return model.get_vector(word)
    else:
        return np.zeros(300)

# Load your word embedding model (e.g., Word2Vec, GloVe, FastText)
# Example using Gensim's pre-trained Word2Vec model
model_path = 'C:\\Users\\user\\Desktop\\Search\\GoogleNews-vectors-negative300.bin'  # Replace with your model path
model = KeyedVectors.load_word2vec_format(model_path, binary=True)

# Sample documents
fin = []

# Calculate average vector for each document
out_dict = {}
for sen in fin:
    average_vector = np.mean(
        np.array([get_embedding(x) for x in word_tokenize(remove_stopwords(sen))]),
        axis=0
    )
    out_dict[sen] = average_vector

# Function to calculate the similarity between the query vector and document vector
def get_sim(query_embedding, average_vector_doc):
    sim = 1 - spatial.distance.cosine(query_embedding, average_vector_doc)
    return sim

# Function to rank documents based on similarity to the query
def Ranked_documents(query):
    query_words = np.mean(
        np.array([get_embedding(x) for x in word_tokenize(query.lower())], dtype=float),
        axis=0
    )
    rank = [(k, get_sim(query_words, v)) for k, v in out_dict.items()]
    rank = sorted(rank, key=lambda t: t[1], reverse=True)
    
    dd = pd.DataFrame(rank, columns=['Desc', 'score'])
    rankfin = pd.merge(data1, dd, left_on='description', right_on='Desc')
    rankfin = rankfin[['product_name', 'description', 'score']]
    print('Ranked Documents:')
    return rankfin

# Assuming 'data1' is a DataFrame with columns 'product_name' and 'description'
data1 = pd.DataFrame({
    'product_name': ['Product1', 'Product2', 'Product3'],
    'description': ["This is the first document.", "This document is the second document.", "And this is the third one."]
})

query = input("What would you like to research? ")
ranked_documents = Ranked_documents(query)
print(ranked_documents)
```
